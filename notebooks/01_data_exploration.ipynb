```python
# Option A (preferred): import from package
from src import data_loader as dl
# Option B: add src/ to sys.path so `import data_loader` works
import sys
from pathlib import Path
ROOT = Path().resolve().parent
sys.path.append(str(ROOT / "src"))

import data_loader as dl
import matplotlib.pyplot as plt

# 1. Load
sig = dl.load_signal('data/raw/yes_ABD.npy')
rms = dl.get_rms(sig)
peaks, thresh = dl.find_event_indices(rms)

# 2. Sanity Check Plot
plt.figure(figsize=(15, 5))
plt.plot(sig, alpha=0.5, label='ABD Raw')
plt.plot(rms, label='Envelope')
plt.axhline(thresh, color='r', ls='--')
plt.scatter(peaks, rms[peaks], color='red', zorder=3)
plt.title(f"Detected {len(peaks)} Events")
plt.show()

# 3. Create Dataset
X, y = dl.build_labeled_dataset(sig, peaks)
print(f"Dataset shape: {X.shape}, Labels: {y}")
```
